```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
from skimage import data
```

```python
def mostrar(imagen):
    plt.imshow(imagen, cmap="gray")
    plt.axis("off")
    plt.show()
```

```python
img_placa = cv.imread('placa.png', 0)
```

```python
mostrar(img_placa)
```

### * Los elementos estructurales son matrices que identifica un pixel para procesar las imagenes

```python
element_structural = cv.getStructuringElement(cv.MORPH_ELLIPSE, (7,7))
```

```python
element_structural
```

### * Las operacione morfologicas son operaciones para procesar imagenes basadas en sus formas
###   done cada pixel de la imagen se ajusta de acuerdo a valores de pixeles vecinos, ejemplos son:
###   Erocion, dilatacion, abertura, gradiente 

```python
# operacion morfologica de dilatacion
img_placa_dilatada = cv.dilate(img_placa, element_structural, iterations=2)
```

```python
mostrar(img_placa_dilatada)
```

```python
element_structural_2 = cv.getStructuringElement(cv.MORPH_RECT, (7,7))
```

```python
element_structural_2
```

```python
# operacion mrfologica de erosion
img_placa_erosionada = cv.erode(img_placa, element_structural_2, iterations=2)
```

```python
mostrar(img_placa_erosionada)
```

```python
img_moneda = cv.imread('moneda.png')
```

```python
mostrar(img_moneda)
```

```python
element_structural_3 = cv.getStructuringElement(cv.MORPH_ELLIPSE, (5,5))
```

```python
element_structural_4 = cv.getStructuringElement(cv.MORPH_ELLIPSE, (21,21))
```

```python
img_moneda_gradiente = cv.morphologyEx(img_moneda, cv.MORPH_GRADIENT, element_structural_3)
```

```python
mostrar(img_moneda_gradiente)
```

### * imagen con estructura morfologica black hat, se le dice morfologicamente abierta

```python
img_moneda_tophat = cv.morphologyEx(img_moneda, cv.MORPH_TOPHAT, element_structural_4)
```

```python
mostrar(img_moneda_tophat)
```

### * imagen con estructura morfologica black hat, se le dice morfologicamente cerrada

```python
img_blackhat = cv.morphologyEx(img_moneda, cv.MORPH_BLACKHAT, element_structural_4)
```

```python
mostrar(img_blackhat)
```
