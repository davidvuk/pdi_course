```python
img_contraste=cv.imread('einstein.jpg',0)
```

```python
plt.imshow(img_contraste,cmap='gray')
plt.axis('off')
plt.show()
```

```python
img_equal=cv.equalizeHist(img_contraste)
```

```python
def mostrar_imagen(imagen):
  plt.imshow(imagen,cmap='gray')
  plt.axis('off')
  plt.show()
```

```python
mostrar_imagen(img_equal)
```

```python
fig=plt.figure(figsize=(20,5))
ax1=fig.add_subplot(1,2,1)
plt.hist(img_contraste.ravel(),bins=256)

ax2=fig.add_subplot(1,2,2)
plt.hist(img_equal.ravel(),bins=256)

plt.show()
```

```python
img_contraste.shape
```

```python
cantFilas,cantColumns=img_contraste.shape
```

```python
cantColumns
```

```python
rot=cv.getRotationMatrix2D((cantFilas/2,cantColumns/2),45,1)
```

```python
img_rot=cv.warpAffine(img_contraste,rot,(cantColumns,cantFilas))
```

```python
mostrar_imagen(img_rot)
```

```python
tra=np.float32([[1,0,100],[0,1,100]])
```

```python
img_tra=cv.warpAffine(img_contraste,tra,(cantColumns,cantFilas))
```

```python
mostrar_imagen(img_tra)
```

```python
img_modif=cv.resize(img_contraste,None,fx=0.8,fy=0.8,interpolation=cv.INTER_CUBIC)
```

```python
mostrar_imagen(img_modif)
```

```python
img1=cv.imread('batman.png',0)
img2=cv.imread('superman.png',0)
```

```python
mostrar_imagen(img1)
```

```python
img_suma=cv.add(img1,img2)
```

```python
mostrar_imagen(img_suma)
```

```python
img_resta=cv.subtract(img2,img1)
```

```python
mostrar_imagen(img_resta)
```

```python
img_result=cv.addWeighted(img1,0.3,img2,0.8,0)
```

```python
mostrar_imagen(img_result)
```
