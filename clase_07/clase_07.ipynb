```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
from skimage import data
```

```python
def mostrar(imagen):
    plt.imshow(cv.cvtColor(imagen, cv.COLOR_BGR2RGB))
    plt.axis("off")
    plt.show()
```

```python
clasificador = cv.CascadeClassifier('cascades/haarcascade_frontalface_default.xml')
```

```python
imagen = cv.imread('personas/personas3.jpg')
```

```python
mostrar(imagen)
```

```python
imagen_gris = cv.cvtColor(imagen, cv.COLOR_BGR2GRAY)
```

```python
mostrar(imagen_gris)
```

```python
rosotros_detectador = clasificador.detectMultiScale(imagen_gris, scaleFactor=1.1, minNeighbors=9)
```

```python
len(rosotros_detectador)
```

```python
for x, y, l, a in rosotros_detectador:
    imagen = cv.rectangle(imagen, (x, y), (x+l, y+a), (0,0,255), 2)
```

```python
mostrar(imagen)
```

```python
import cv2 as cv

video = cv.VideoCapture(0)
clasificador1 = cv.CascadeClassifier('haarcascade_frontalcatface_default.xml')

while True:
    conectado, frame = video.read()

    frame_gris = cv.cvtColor(frame, cv.COLOR_BGR2GRAY)

    rosotros_detectados = clasificador1.detectMultiScale(frame_gris)

    for(x,y,l,a) in rosotros_detectados:
        cv.rectangle(frame, (x,y), (x+l,y+a), (0,0,255), 2)
    
    cv.imshow('Video', frame)

    if cv.waitKey(1) == ord('q'):
        break

video.release()
cv.destroyAllWindows()
```
